## Libraries and packages used

```python
from amnesiac import blurry_memory
import random
import string
import random
import numpy as np
import matplotlib.pyplot as plt
import time
from random import randint
```

## Functions

```python
def new_population(population_size):
    population = []
    for i in range(population_size):
#         chromosome = []
#         for x in range(passcode_length):
#             chromosome.append(randomStringDigits(passcode_length))
        population.append(randomStringDigits(passcode_length))
        
    return population
```

```python
def randomStringDigits(stringLength):
    
    """Generate a random string of letters and digits """
    lettersAndDigits = string.ascii_uppercase + string.digits +'_'
    
    return ''.join(random.choice(lettersAndDigits) for i in range(stringLength))
```

```python
def sort_population(fitness_scores):
    parents_list = []
    for chromosome in sorted(fitness_scores, key=lambda x: x[1], reverse = True)[:num_parents]:
        parents_list.append(chromosome[0])

        
    return(parents_list)
```

```python
def get_fitness(population, target):
    
#     print('population before fitness',population)

    pop_scores={}

    for i in range(len(population)-1):
        pop_scores.update(blurry_memory((population[i], population[1+i]), 181004877, target))
    #print(test)

    scores = [ [k,v] for k, v in pop_scores.items() ]
    
#     print('population after fitness',scores)

    return scores
```

```python
def selection(pop_sort):
    
    # New population
    children = []
    
    # Elitism
    if selection_type == 'elite':
        
        for i in range(elite_size):
            children.append(pop_sort[i])
        
        parent1 = 0
        while(len(children)<population_size):
            parent2 = randint(0, population_size-1)
            while parent1 == parent2:
                 parent2 = randint(0, population_size-1)
#             print('CHECK', parent1, parent2)
            #print(pop_sort)
            children.append(crossover(pop_sort[parent1], pop_sort[parent2]))
            parent1 += 1
            
        
    return children
```

```python
def crossover(parent1, parent2):
    
    child = []
    
    if cross_type == 'one_point':
            
        cross_point = 5 #----------------------------------------------
        child[:cross_point] = parent1[:cross_point]
        child[cross_point:] = parent2[cross_point:]
        
    return ''.join(child)
```

```python
def mutation(population):
    
    for i in range(len(population)):
        
        if random.random() < mutation_rate:
            
            if mutation_type == 'bit_flip':
                position = randint(0, passcode_length-1)
#                 print('\nBef', population[i], 'pos:', position)
                temp_list = list(population[i])
                temp_list[position] = randomStringDigits(1)
                population[i] = ''.join(temp_list)
#                 print('Aft', population[i])
                
    return population
```

## Main Coding

```python
#Parameters
target = 0
max_generetations = 100000
passcode_length = 10
population_size = 50
num_parents = population_size
elite_size = 2
mutation_rate = 0.2

disp = True

selection_type = 'elite'
cross_type = 'one_point'
mutation_type = 'bit_flip'
```

### Main Loop

```python
population = new_population(population_size)

stop = False
count = 0
best_ind = ''
best_fit = 0

while stop==False and count <= max_generetations:
    
    # Get score for each individual
    pop_scores = get_fitness(population, target)
#     if disp:
#         print('After fitness', pop_scores, '\n')
        
    # Stop condition
    for i in range(len(pop_scores)):
        if pop_scores[i][1] > best_fit:
            best_fit = pop_scores[i][1]
            best_ind = pop_scores[i][0]
        
        if pop_scores[i][1]==1.0:
            winner_i = pop_scores[i][0]
            # PASSWORD FOUND
            stop=True

        
    # Sort population
    pop_sorted = sort_population(pop_scores)
#     if disp:
#         print('After sorting', pop_sorted, '\n')
        
    # Sometimes....
    while len(pop_sorted) < population_size:
        pop_sorted.append(randomStringDigits(10))
    
    
    # Selection and crossover
    offspring = selection(pop_sorted)
#     if disp:
#         print('After selec. cross.', offspring, '\n')
        
    # Mutation
    n_population = mutation(offspring)
#     if disp:
#         print('After mutation', n_population, '\n')
        
    # Restart
    population = n_population
            
    count += 1
    
    # Visualization
    if disp:
        if count % 500 == 0:
            print('\n~~~~~~~~~~~~~~~~~~~')
            print('~~~~~~SUMMARY~~~~~~')
            print('~~~~~~~~~~~~~~~~~~~')
            print('GENERATION: ', count)
            print('Highest fitness:', best_fit)
            print('Best individual:', best_ind)
            print('~~~~~~~~~~~~~~~~~~~')
            print('Current population:')
            print(n_population)
            
    
print('\n----------------------')
print('PASSWORD:', winner_i)
#print(count)
```
